### Описание библиотеки

Для описания выбрал Mystem от Яндекса. Как оказалось, у него закрытый исходный код и он поддерживает только русский язык. Запускать ее можно и как консольное приложение. Судя по отчету прошлых годов работает дольше чем другие анализаторы, зато проводит лемматизацию, а не стемминг, основываясь на дереве суффиксов. Использует словарь НКРЯ (Национальный Корпус Русского Языка), в котором собрано более 260 тысяч лемм.

```python
from pymystem3 import Mystem
import time
mystem = Mystem()
```

### Тесты

Попробую проанализировать текст на ~1400 слов. Возьму первые 10 слов и попробую проанализировать их по отдельности (заранее знаю, что выйдет долго, поэтому беру только 10)

```python
start_time = time.time()
raw_lemmas_word_by_word = list(
    map(
        mystem.analyze,
        text.split()[:10]
    )
)
"--- %s seconds ---" % (time.time() - start_time)
```

У меня получилось 7 секунд для 10 слов - очень долго. Попробую дать на вход сразу весь текст.

```python
raw_lemmas_word_by_word
```

```python
start_time = time.time()
raw_lemmas = mystem.analyze(text)
"--- %s seconds ---" % (time.time() - start_time)
```

С другой стороны текст целиком обрабатывается меньше секунды. Т.е. лучше подавать на вход сразу весь текст.

```python
raw_lemmas
```

Существенных различий в двух способах обработки нет.

При этом сам разбивает текст на слова, хотя пробелы и другие символы не удаляет, приходится удалять самим.

```python
def has_letters(s: str):
    return any(str(c).isalpha() for c in s)
```

```python
words_lemmas = list(
    filter(
        lambda x: has_letters(x.get('text')),
        raw_lemmas
    )
)
words_lemmas
```
